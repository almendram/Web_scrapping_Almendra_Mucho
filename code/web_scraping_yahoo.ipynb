```python
# Importar herramientas de Selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from selenium.webdriver import ActionChains
import yfinance as yf

# WebDriver Manager
from webdriver_manager.chrome import ChromeDriverManager

# Standard libraries
import pandas as pd
import numpy as np
import os
import time
import re
```

```python
# Configuraci贸n de la inciaci贸n del navegador
chrome_options = Options()
# chrome_options.add_argument("--headless")
chrome_options.add_argument("--start-maximized")
chrome_options.add_argument("--lang=en-US")
# chrome_options.add_argument("--disable-gpu")
# chrome_options.add_argument("--no-sandbox")
# chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--headless")

# Iniciar el WebDriver de Chrome
driver = webdriver.Chrome()
# pasando nuestras 'chrome_options' como argumento.

print("WebDriver iniciado con 茅xito.")
```

```python
# Extracci贸n de datos
options.add_argument("--headless")  
driver = webdriver.Chrome(options=options)

url = "https://finance.yahoo.com/markets/stocks/gainers/?start=0&count=50"
driver.get(url)

table = WebDriverWait(driver, 15).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, "table"))
)

# Seleccionamos desde la tabla
row_selector = "table tbody tr"
rows = driver.find_elements(By.CSS_SELECTOR, row_selector)

stocks_data = []

for row in rows[:50]:
    try:
        symbol = row.find_element(By.CSS_SELECTOR, "td:nth-child(1) a").text
        name = row.find_element(By.CSS_SELECTOR, "td:nth-child(2)").text
        
        stocks_data.append({
            "Simbolo": symbol,
            "Nombre": name
        })
        
        print(f"Scraped: {symbol} - {name}")

    except Exception as e:
        print(f"Error extrayendo datos de una acci贸n: {e}")
        continue

print(f"\nScraping completado. Se extrajeron {len(stocks_data)} acciones.")
driver.quit()
```

```python
# Se guarda dataframe
if stocks_data:
    df = pd.DataFrame(stocks_data)
    
    df.to_csv(r"..\outputs\Stocks_50_data.csv", index=False, encoding='utf-8-sig')
    print("\nDatos guardados exitosamente.")
    
    display(df.head())
else:
    print("\nNo se pudo extraer ning煤n dato de las acciones.")

driver.quit()
print("\nNavegador cerrado correctamente. 隆Ejercicio terminado!")
```

```python
# Se extraen los datos
simbolos = [stock["Simbolo"] for stock in stocks_data]

data = yf.download(
    tickers=simbolos,
    period="1y",
    interval="1mo",
    group_by="ticker",
    auto_adjust=True
)

precios_cierre = pd.DataFrame({
    simbolo: data[simbolo]["Close"] for simbolo in simbolos
})
print(precios_cierre)
```

```python
#### Parte 3
```

```python
# C谩lculo de retornos mensuales (primeros 6 meses) y rentabilidad acumulada
precios_cierre = precios_cierre.sort_index()

primeros6 = precios_cierre.iloc[:6]
ultimos6 = precios_cierre.iloc[6:]

retornos_6 = (primeros6 / primeros6.shift(1) - 1).dropna()
rentabilidad_6 = (1 + retornos_6).cumprod() - 1

volatilidad_6 = retornos_6.std() # Volatilidad

# Selecci贸n de la cartera: Se optar谩 por escoger a las 4 acciones con mayor tasa de retorno los pr贸ximos 6 meses
# y a las 6 acciones con menor volatilidad registrada en los primeros 6 meses.

top4 = rentabilidad_6.sort_values(ascending=False).head(4).index.tolist()
resto = [s for s in retornos_6.columns if s not in top4]
lowvol6 = volatilidad_6.loc[resto].sort_values().head(6).index.tolist()

cartera = top4 + lowvol6
print("Cartera seleccionada:", cartera)
```

```python
# C谩lculo de rendimiento porcentual mensual de los 煤ltimos 6 meses de las acciones
ultimos6_sel = ultimos6[cartera]
retornos_acciones = (ultimos6_sel / ultimos6_sel.shift(1) - 1).dropna()
print("\n Rendimientos mensuales de las 10 acciones seleccionadas:")
display(retornos_acciones)
```

```python
# Rendimientos de la cartera (ponderaciones iguales en 10%)
retornos_cartera = retornos_acciones.mean(axis=1)

print("\n Rendimientos mensuales de la cartera (10 acciones):")
display(retornos_cartera)

retornos_acciones.to_csv("../outputs/Rendimientos_Acciones.csv", encoding="utf-8-sig")
```
